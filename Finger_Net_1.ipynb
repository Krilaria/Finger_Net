Загрузка данных из датасета

```python
import json
import random
import time
```

```python
file_path = '/content/events_008.txt'

with open(file_path, 'r') as file:
    content = file.read()

print(content)
```

```python
# Декодирование JSON
data = json.loads(content)

# Извлечение информации
interval = data['interval']
duration = data['duration']
client_info = data['clientInfo']
mouse_move_events = data['events']['mouseMove']
click_events = data['events']['clicks']
scroll_events = data['events']['scrolls']
tap_events = data['events']['taps']
touch_events = data['events']['touches']

# Вывод информации
print("Interval:", interval)
print("Duration:", duration)
print("Client Info:", client_info)
print("Mouse Move Events:", mouse_move_events)
print("Click Events:", click_events)
print("Scroll Events:", scroll_events)
print("Tap Events:", tap_events)
print("Touch Events:", touch_events)
```

Генерация отчёта со случайными движениями мышкой

```python
def modify_coordinates_1(input_file, output_file):
    with open(input_file, 'r') as f:
        data = json.load(f)

    # Здесь вы можете изменить координаты или выполнять другие манипуляции с данными
    # Например, заменить координаты на случайные значения или изменить существующие

    for event in data['events']['mouseMove']:
        event['left'] = random.randint(0, 200)
        event['top'] = random.randint(0, 200)

    # Сохраняем измененные данные в новый файл
    with open(output_file, 'w') as f:
        json.dump(data, f)
```

```python
def modify_coordinates_2(input_file, output_file, max_displacement=10):
    with open(input_file, 'r') as f:
        data = json.load(f)
    last_x, last_y = None, None
 # Более плавные движения мышкой
    for event in data['events']['mouseMove']:
        if last_x is not None and last_y is not None:
            t = random.random()  # Случайное значение t от 0 до 1
            event['left'] = last_x + t * (event['left'] - last_x)
            event['top'] = last_y + t * (event['top'] - last_y)

        # Обновляем координаты для следующей итерации
        last_x, last_y = event['left'], event['top']

        # Добавляем случайное смещение к новым координатам
        event['left'] += random.randint(-max_displacement, max_displacement)
        event['top'] += random.randint(-max_displacement, max_displacement)

    # Сохраняем измененные данные в новый файл
    with open(output_file, 'w') as f:
        json.dump(data, f)
```

```python
def modify_coordinates_3(input_file, output_file, max_displacement=5, pause_time=0.02):
    with open(input_file, 'r') as f:
        data = json.load(f)
    last_x, last_y = None, None

    for event in data['events']['mouseMove']:
        if last_x is not None and last_y is not None:
            t = random.random()  # Случайное значение t от 0 до 1
            event['left'] = last_x + t * (event['left'] - last_x)
            event['top'] = last_y + t * (event['top'] - last_y)

        # Обновляем координаты для следующей итерации
        last_x, last_y = event['left'], event['top']

        # Добавляем случайное смещение к новым координатам
        event['left'] += random.randint(-max_displacement, max_displacement)
        event['top'] += random.randint(-max_displacement, max_displacement)

        # Пауза между движениями
        time.sleep(pause_time)

    # Сохраняем измененные данные в новый файл
    with open(output_file, 'w') as f:
        json.dump(data, f)
```

```python
# Пример вызова функции
input_file = 'events_008.txt'
output_file = 'script_008.txt'
modify_coordinates_3(input_file, output_file)
```
